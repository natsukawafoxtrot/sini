```python
import random

def generate_maze(rows, cols, obstacles):
    maze = [[' ' for _ in range(cols)] for _ in range(rows)]
    start_row = random.randint(0, rows - 1)
    start_col = random.randint(0, cols - 1)
    maze[start_row][start_col] = 'S'
    end_row = random.randint(0, rows - 1)
    end_col = random.randint(0, cols - 1)
    maze[end_row][end_col] = 'E'

    for _ in range(obstacles):
        obstacle_row = random.randint(0, rows - 1)
        obstacle_col = random.randint(0, cols - 1)
        while maze[obstacle_row][obstacle_col] != ' ':
            obstacle_row = random.randint(0, rows - 1)
            obstacle_col = random.randint(0, cols - 1)
        maze[obstacle_row][obstacle_col] = '#'

    return maze

def print_maze(maze):
    rows = len(maze)
    cols = len(maze[0])

    print("+" + "---+" * cols)
    for i in range(rows):
        row = "|"
        for j in range(cols):
            row += " {} |".format(maze[i][j])
        print(row)
        print("+" + "---+" * cols)

def play_game(maze):
    rows = len(maze)
    cols = len(maze[0])
    current_row = None
    current_col = None

    for i in range(rows):
        for j in range(cols):
            if maze[i][j] == 'S':
                current_row = i
                current_col = j

    while True:
        print_maze(maze)
        move = input("Enter 'U' (up), 'D' (down), 'L' (left), or 'R' (right) to move: ")

        if move == 'U':
            if current_row > 0 and maze[current_row - 1][current_col] != '#':
                current_row -= 1
        elif move == 'D':
            if current_row < rows - 1 and maze[current_row + 1][current_col] != '#':
                current_row += 1
        elif move == 'L':
            if current_col > 0 and maze[current_row][current_col - 1] != '#':
                current_col -= 1
        elif move == 'R':
            if current_col < cols - 1 and maze[current_row][current_col + 1] != '#':
                current_col += 1
        else:
            print("Invalid move! Try again.")
            continue

        if maze[current_row][current_col] == 'E':
            print("Congratulations, you've reached the end!")
            break

        maze[current_row][current_col] = 'X'

# Main program
print("Welcome to Math Maze!")
print("Find your way from 'S' (start) to 'E' (end) while avoiding walls '#' and obstacles.")
print("Enter 'U' (up), 'D' (down), 'L' (left), or 'R' (right) to move.")
print()

maze = generate_maze(5, 5, 3)  # Customize the maze size and number of obstacles here
play_game(maze)
```
